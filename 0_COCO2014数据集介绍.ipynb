## 下载后的数据集应包含下列文件：
- ### train2014: 82783个图片文件
 - ### val2014:  40504个图片文件
 - ### test2014: 40775个图片文件
 - ### annotations: # 包含训练集和验证集的标注文件
     - ### caption_train2014.json: 训练集图像描述的标注文件
     - ### instances_train2014.json: 训练集目标检测与实例分割的标注文件
     - ### person_keypoint_train2014.json: 训练集人体关键点的标注文件
     - ### caption_train2014.json: 验证集图像描述的标注文件
     - ### instances_train2014.json: 验证集目标检测与实例分割的标注文件
     - ### person_keypoint_train2014.json: 验证集人体关键点的标注文件


## json文件的描述：
- ### 每个json文件均包含info、images、licenses、annotations、categories这5个键值对,其中info、images、licenses是所有json文件一样的，annotations、categories根据标注类型不同而不同

- ### info键值对：介绍数据集基本信息

- ### images键值对：每张图片详细信息，并且id为唯一的

- ### licenses键值对：包括下图中的内容，里面集合了不同类型的licenses，并在images中按照id号被引用，基本不参与到数据解析过程中。

- ### annotations键值对：包括下图中的内容，每个序号对应一个注释，一张图片上可能有多个注释。
    - category_id：该注释的类别id；
    - id：当前注释的id号
    - image_id：该注释所在的图片id号
    - area：区域面积bbox：目标的矩形标注框
    - iscrowd：0或1。0表示标注的单个对象，此时segmentation使用polygon表示；1表示标注的是一组对象，此时segmentation使用RLE格式。segmentation：
        - 若使用polygon标注时，则记录的是多边形的坐标点，连续两个数值表示一个点的坐标位置，因此此时点的数量为偶数若使用RLE格式（Run Length Encoding（行程长度压缩算法））
